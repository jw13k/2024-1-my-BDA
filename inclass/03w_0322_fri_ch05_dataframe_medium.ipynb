##05데이터분석기초 데이터 파악하기, 다루기 쉽게 수정하기

|함수|기능|
|----|----|
head()|앞부분 출력|
tail()|뒷부분 출력|
shape|행, 열 개수 출력|
info()|변수 속성 출력|
